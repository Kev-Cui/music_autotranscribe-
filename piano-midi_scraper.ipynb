```python
import os
import requests
from bs4 import BeautifulSoup
import time
```

```python
# This code gets all composers' link from the database

url = "http://www.piano-midi.de/midi_files.htm"
response = requests.get(url)

if response.status_code == 200: # if successful
    soup = BeautifulSoup(response.content, 'html.parser')
    midi_td_elements = soup.find_all('td', class_='midi') # find all <td> elements with class "midi"
    
    midi_links = [] # list to store extracted links
    
    # Extract links from <a> tags within <td class="midi">
    for td_element in midi_td_elements:
        a_tags = td_element.find_all('a')
        for a_tag in a_tags:
            link = a_tag.get('href')
            midi_links.append(link)
else:
    print("Failed to fetch the webpage. Status code:", response.status_code)

# prefix to make direct-access links
prefixed_links = ['http://www.piano-midi.de/' + link for link in midi_links]
prefixed_links
```

```python
# This code gets all download links of wanted files

# Initiate list to store download links for MIDI and MP3 files
midi_download_links = []
mp3_download_links = []

# Loop through prefixed_links
for link in prefixed_links:
    response = requests.get(link)
    if response.status_code == 200: # if successful
        soup = BeautifulSoup(response.content, 'html.parser')
        
        midi_td_elements = soup.find_all('td', class_='midi')# find all <td> elements with class "midi"

        # Extract download links for MIDI files
        for td_element in midi_td_elements:
            a_tags = td_element.find_all('a', href=lambda href: href and href.startswith('midis/'))
            for a_tag in a_tags:
                midi_link = a_tag.get('href')
                midi_download_links.append(midi_link)
        
        # Extract download links for MP3 files
        mp3_a_tags = soup.find_all('a', href=lambda href: href and href.startswith('mp3/'))
        for mp3_a_tag in mp3_a_tags:
            mp3_link = mp3_a_tag.get('href')
            mp3_download_links.append(mp3_link)
    else:
        print(f"Failed to fetch the webpage {link}. Status code:", response.status_code)

# prefix the list of download links
prefixed_midi_links = ['http://www.piano-midi.de/' + link for link in midi_download_links]
print(prefixed_midi_links)
prefixed_mp3_links = ['http://www.piano-midi.de/' + link for link in mp3_download_links]
print(prefixed_mp3_links)
```

```python
# I want format 0 midi files, so remove all midi links that does not contain 'format0'
print('before removal:',len(prefixed_midi_links))
prefixed_midi_links = [link for link in prefixed_midi_links if '_format0' in link]
print('after removal:',len(prefixed_midi_links))
prefixed_midi_links
```

```python
print(len(prefixed_mp3_links)) # will remove extras
```

```python
def extract_midi_name(url): # gets midi file name
    return url.split('/')[-1].split('_format0')[0]
def extract_mp3_name(url): # gets mp3 file name
    return url.split('/')[-1].split('.mp3')[0]

midi_file_names = [extract_midi_name(midi) for midi in prefixed_midi_links]

filtered_mp3_links = [mp3 for mp3 in prefixed_mp3_links if extract_mp3_name(mp3) in midi_file_names]
len(filtered_mp3_links)
```

```python
# Download function
def download_files(download_links, folder_name="default_folder"):
    if not os.path.exists(folder_name): #creates folder if not exsist in the working directory
        os.makedirs(folder_name)

    for link in download_links:
        try:
            file_name = link.split('/')[-1]
            save_path = os.path.join(folder_name, file_name)

            # Download and save
            response = requests.get(link)
            if response.status_code == 200:
                with open(save_path, 'wb') as f:
                    f.write(response.content)
                print(f"Downloaded {file_name} successfully.")
            else:
                print(f"Failed to download {file_name}. Status code: {response.status_code}")
            time.sleep(0.5)
        except Exception as e:
            print(f"Error downloading {file_name}: {str(e)}")
```

```python
# I download
download_files(prefixed_midi_links, folder_name="piano_midi")
download_files(filtered_mp3_links, folder_name="piano_mp3")
```
